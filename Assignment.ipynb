```python
from datetime import timedelta, date
from xml.dom.minidom import parse, parseString 
import xml.etree.ElementTree as ET
import os
def praseXml(x,y):
    fpath = os.path.join("/home/acer/Downloads/proxy/Python_exercises_QA_Engr/Updated_Python_exercises_QA_Engr", "test_payload1.xml")
    f = open(fpath)
    content = f.read()
    
    file = xml.dom.minidom.parse(fpath)
  
    update_depart_date=date.today() + timedelta(days=x)
    update_return_date=date.today() + timedelta(days=y)
    update_depart_date=update_depart_date.strftime("%Y")+update_depart_date.strftime("%m")+update_depart_date.strftime("%d")
    update_depart_date=update_return_date.strftime("%Y")+update_return_date.strftime("%m")+update_return_date.strftime("%d")
    delete=file.getElementsByTagName('DEPART')
    for i in delete: 
        x = i.parentNode
        x.removeChild( i )
    
    current_Return_date=file.getElementsByTagName('RETURN')
    for i in  current_Return_date:   
        x = i.parentNode
        x.removeChild( i )
    tag=file.getElementsByTagName('TP')
    DEPART = file.createElement( "DEPART" )
    
    tag[0].appendChild(DEPART)
    DEPART.setAttribute("value", update_depart_date)
    
    RETURN = file.createElement( "RETURN" )
    tag[0].appendChild(RETURN)
    RETURN.setAttribute("value", update_return_date) 
    #print(update_depart_date)
    

praseXml(4,7)
```
